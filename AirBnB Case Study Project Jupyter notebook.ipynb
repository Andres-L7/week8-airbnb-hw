```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../files/AB_NYC_2019.csv')
air_bnb.head(20)
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.groupby('neighbourhood_group').count()

# 5 neighborhood groups, Manhattan 
```

```python
# Are private rooms the most popular in manhattan?
air_bnb.groupby(['room_type', 'neighbourhood_group']).count()

#No, entire homes are most popular
```

```python
# Which hosts are the busiest and based on their reviews?
air_bnb.groupby('host_id').sum()[['id', 'number_of_reviews']].sort_values('number_of_reviews', ascending = False)

# host id 37312959
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby('neighbourhood_group').mean()

#Manhattan
```

```python
# Which neighborhood group has the highest total price?
air_bnb.groupby('neighbourhood_group').sum()

#Manhattan
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby('host_id').sum()[['id','price']].sort_values('price', ascending = False)

# host ids: 219517861,107434423,156158778,205031545,30283594
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
not_avail = air_bnb[air_bnb['availability_365'] == 0]
not_avail[not_avail['number_of_reviews'] >= 100]

# it says "162 rows", so im guessing those are the ones unavailible
```

```python
# What host has the highest total of prices and where are they located?
air_bnb.groupby('host_id').sum()[['id','price']].sort_values('price', ascending = False)
# Host ID: 219517861

air_bnb[air_bnb['host_id']==219517861]
#their location is manhattan
```

```python
# When did Danielle from Queens last receive a review?
danielle_bnb = air_bnb[air_bnb['host_name']=='Danielle']
danielle_bnb[danielle_bnb['neighbourhood_group']=='Queens'][['host_name','last_review']].sort_values('last_review',ascending=False)

# it shows last review: 2019-07-08
```
